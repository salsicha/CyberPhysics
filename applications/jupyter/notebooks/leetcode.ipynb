```python
# Amazon question

# Find the pair of indices in Arg1 that sum to Arg2.

arg1 = [5, 3, 6, 1, 9]
arg2 = -3

# Answer: Binary search

import bisect
for x in arg1:
  diff = x - arg2
  index = bisect.bisect_left(arg1, diff)
  if index == len(arg1):
    # return False
      print("fail")
      break
  elif index + 1 == diff:
    # return True
      print("success")
      break
print("error") # log error
# return False

```
